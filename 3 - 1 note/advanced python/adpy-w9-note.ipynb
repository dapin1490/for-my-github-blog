### 예제

```python
import math as m

x = range(0, 10)
y1 = [m.cos(v) for v in x]
y2 = [m.sin(v) for v in x]

x1 = range(0, 3)
z1 = [m.exp(v) for v in x1]
z2 = [m.exp(2*v) for v in x1]

plt.figure(figsize=(10, 9))

plt.subplot(2, 1, 1)
plt.plot(x, y1, "--rd")  # cos
plt.plot(x, y2, ":bx")  # sin
plt.xlabel("x")
plt.ylabel("cos(x) and sin(x)")
plt.title("Trigonometric functions")
plt.legend(["cos(x)", "sin(x)"], loc="lower left")

plt.subplot(2, 1, 2)
plt.plot(x1, z1, "-.ko")  # exp(x)
plt.plot(x1, z2, "-m^")  # exp(2x)
plt.xlabel("x")
plt.ylabel("exp(x) and exp(2x)")
plt.title("Exponential functions")
plt.legend(["exp(x)", "exp(2x)"], loc="upper left")

plt.show()
```

### 답안

```python
import matplotlib.pyplot as plt
import math as m

x = range(0,10)
y1 = [m.cos(v) for v in x]
y2 = [m.sin(v) for v in x]

x1 = range(0,3)
z1 = [m.exp(v) for v in x1]
z2 = [m.exp(2*v) for v in x1]

plt.subplot(2,1,1) # row의 개수, column의 개수, 몇 번째
plt.title('Trigonometric functions')
plt.ylabel('cos(x) and sin(x)')
plt.xlabel('x')
plt.plot(x,y1,'--rd')
plt.plot(x,y2,':bx')
plt.legend(['cos(x)','sin(x)'])

plt.subplot(2,1,2)
plt.title('Exponential functions')
plt.ylabel('exp(x) and exp(2x)')
plt.xlabel('x')
plt.plot(x1,z1,'-.ko')
plt.plot(x1,z2,'-m^')
plt.legend(['exp(x)','exp(2x)'])

plt.show()
```
