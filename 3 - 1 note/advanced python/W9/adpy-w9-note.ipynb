## plot : 기본 선그래프

```python
import matplotlib.pyplot as plt

x = (-4, -3, -2, -1, 0, 1, 2, 3, 4)
y = (16, 9, 4, 1, 0, 1, 4, 9, 16)
plt.plot(x, y)
plt.title("Plot Example")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.show()
```

## subplot(), figure size   
   
인덱스 순서 : 1부터 시작, 왼쪽 상단부터 오른쪽으로 진행하며 가로줄을 다 센 후 아래로 내려감.   

```python
import matplotlib.pyplot as plt

x = range(0, 100)
y = [v for v in x]
y1 = [v ** 2 for v in x]
y2 = [v ** 3 for v in x]

plt.figure(figsize=(15, 5))

plt.subplot(2, 1, 1) # row의 개수, column의 개수, 몇 번째
plt.ylabel('v^2')
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.ylabel('v^3')
plt.plot(x, y2)

plt.show()

# ----------------

plt.figure(figsize=(5, 10))

plt.subplot(2, 1, 1) # row의 개수, column의 개수, 몇 번째
plt.ylabel('v^2')
plt.plot(x, y1)

plt.subplot(2, 1, 2)
plt.ylabel('v^3')
plt.plot(x, y2)

plt.show()
```

## 다른 그래프 유형(bar, barh)

### 예제

```python
import matplotlib.pyplot as plt
import math

x = range(0, 10)
y1 = [math.cos(v) for v in x]
y2 = [math.sin(v) for v in x]

x1 = range(0, 3)
z1 = [math.exp(v) for v in x1]
z2 = [math.exp(2*v) for v in x1]

plt.figure(figsize=(7, 5))

plt.subplot(2, 1, 1)
plt.plot(x, y1, "--rd")  # cos
plt.plot(x, y2, ":bx")  # sin
plt.xlabel("x")
plt.ylabel("cos(x) and sin(x)")
plt.title("Trigonometric functions")
plt.legend(["cos(x)", "sin(x)"], loc="lower left")

plt.subplot(2, 1, 2)
plt.plot(x1, z1, "-.ko")  # exp(x)
plt.plot(x1, z2, "-m^")  # exp(2x)
plt.xlabel("x")
plt.ylabel("exp(x) and exp(2x)")
plt.title("Exponential functions")
plt.legend(["exp(x)", "exp(2x)"], loc="upper left")

plt.tight_layout() # 각 그래프간 겹치지 않도록 함

plt.show()
```

### 답안

```python
import matplotlib.pyplot as plt
import math

x = range(0,10)
y1 = [math.cos(v) for v in x]
y2 = [math.sin(v) for v in x]

x1 = range(0,3)
z1 = [math.exp(v) for v in x1]
z2 = [math.exp(2*v) for v in x1]

plt.subplot(2,1,1) # row의 개수, column의 개수, 몇 번째
plt.title('Trigonometric functions')
plt.ylabel('cos(x) and sin(x)')
plt.xlabel('x')
plt.plot(x,y1,'--rd')
plt.plot(x,y2,':bx')
plt.legend(['cos(x)','sin(x)'])

plt.subplot(2,1,2)
plt.title('Exponential functions')
plt.ylabel('exp(x) and exp(2x)')
plt.xlabel('x')
plt.plot(x1,z1,'-.ko')
plt.plot(x1,z2,'-m^')
plt.legend(['exp(x)','exp(2x)'])

plt.show()
```

## 원과 사각형 그리기

```python
import matplotlib.pyplot as plt

plt.axes() # 새로운 축 생성
circle = plt.Circle((-1, -1), 1, fc='g', ec='red')
plt.gca().add_patch(circle) # 모양 축에 더하기
# gca(): 이전 축이 있다면 해당 축을 사용하는 기능
rectangle = plt.Rectangle((0, 0), 4, 2, fc='b', ec='red')
plt.gca().add_patch(rectangle)
plt.axis([-3, 5, -3, 3])
# 그림의 범위 조절 xmin, xmax, ymin, ymax 순서
plt.grid() # linestyle='--'와 같이 linestyle 지정 가능

plt.show()
```

## clip on

```python
import matplotlib.pyplot as plt

circle1 = plt.Circle((-1, -1), 0.8)
circle2 = plt.Circle((0, 0), 0.4)
circle3 = plt.Circle((1, 1), 0.2, clip_on=False)  # clip_on=False : 그림이 그리드 밖으로 빠져나가도 다 그려줌

f, ax = plt.subplots() # f라는 figure, ax라는 axe 생성

plt.xlim(-1, 1) #x축 크기
plt.ylim(-1, 1) #y축 크기
ax.set_aspect(1) # 동일한 축의 정사각형 생성
# 예를 들어 3이면 축의 비율이 달라짐

ax.add_artist(circle1)
ax.add_artist(circle2)
ax.add_artist(circle3)

plt.show()
```

## 객체지향 방식 subplot (fig.add_subplot)

```python
import matplotlib.pyplot as plt
import numpy as np # 차주에 배울 numpy 라이브러리 사용

fig = plt.figure(figsize=(10,3))

t = np.linspace(0.0, 1.0)

for subplot in range(1, 7):
    axis = fig.add_subplot(2, 3, subplot)
    axis.plot(t, np.cos(2 * np.pi * t * subplot) + 1)
    axis.set_xlabel('time[s]')
    axis.set_ylabel('voltage[mV]')
    axis.set_title('$\cos({}\pi t)$ +1'.format(subplot))

fig.tight_layout() # 각 그래프간 겹치지 않도록 함

plt.show()
```

## 그래프 안의 그래프

```python
import matplotlib.pyplot as plt
import numpy as np # 차주에 배울 numpy 라이브러리 사용

fig = plt.figure(figsize=(10, 2))

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.45, 0.6, 0.1, 0.2])

x = np.arange(0, 180, 1) # range와 유사. 차주에 배울 예정
y = np.cos(2 * x * np.pi / 180) + 1

axis1.plot(x, y)
axis2.plot(x, y)

axis2.set_xbound(50, 70)
axis2.set_ybound(0.3, 0.7)

plt.show()
```
